## FLIGHT PREDICATION DATA [EDA & FEATURE ENG.]

### Problem Statement 
The airline industry experiences frequent fluctuations in ticket prices based on multiple factors such as airline, route, travel date, duration, and number of stops. 
For travelers and businesses, understanding the factors influencing flight prices is crucial for cost optimization and planning.

The objective of this project is to perform Exploratory Data Analysis (EDA) and Feature Engineering on flight ticket data to:
- Understand the key factors affecting flight prices
- Identify patterns and trends in pricing behavior
- Transform raw, unstructured features into a clean, numerical, model-ready dataset

This project focuses on data understanding and feature engineering rather than model optimization, making it suitable for analyzing real-world structured data.


## Dataset Overview
The dataset contains flight ticket information for domestic flights in India. 
Each row represents a unique flight journey with details related to the airline, travel schedule, route, and ticket price.

## Fetaure Engineering

#### DATE OF JOURNEY

```python
final_df['Date']=final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month']=final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year']=final_df['Date_of_Journey'].str.split('/').str[2]
```

* Converting  OBJ to INT

```python
final_df['Date']=final_df['Date'].astype(int)
final_df['Month']=final_df['Month'].astype(int)
final_df['Year']=final_df['Year'].astype(int)
```

```python
final_df.drop('Date_of_Journey',axis=1,inplace=True)
```

#### ARRIVAL TIME

```python
final_df['Arrival_Time']=final_df['Arrival_Time'].str.split(' ').str[0]
final_df['Arrival_hour']=final_df['Arrival_Time'].str.split(':').str[0]
final_df['Arrival_min']=final_df['Arrival_Time'].str.split(':').str[1]
```

* Converting OBJ TO INT

```python
final_df['Arrival_hour']=final_df['Arrival_hour'].astype(int)
final_df['Arrival_min']=final_df['Arrival_min'].astype(int)
```

```python
final_df.drop('Arrival_Time',axis=1,inplace=True)
```

#### Departure time

```python
final_df['Dept_hour']=final_df['Dep_Time'].str.split(':').str[0]
final_df['Dept_min']=final_df['Dep_Time'].str.split(':').str[1]
```

* Converting OBJ to INT

```python
final_df['Dept_hour']=final_df['Dept_hour'].astype(int)
final_df['Dept_min']=final_df['Dept_min'].astype(int)
```

```python
final_df.drop('Dep_Time',axis=1,inplace=True)
```

```python
final_df.head()
```

### Total Stops

```python
final_df['Total_Stops'].unique()
```

```python
final_df['Total_Stops']=final_df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'nan':1})
```

```python
final_df.drop('Route',axis=1,inplace=True)
```

## Challenges in the Dataset

* Flight duration is provided in mixed text formats (e.g., "2h 50m")
* Categorical variables have multiple unique categories
* Route and stop information requires transformation
